```python
df['Rating'] = df['Rating'].replace(-1, np.nan)
df['Founded'] = df['Founded'].replace(-1, np.nan)
```

```python
#clean salary

def clean_salary(salary):

    if salary == '-1' or pd.isnull(salary):
        return None, None

    salary = salary.replace('K', '000')
    salary = salary.replace('$', '')
    salary = salary.replace('(Glassdoor est.)', '')
    salary = salary.strip()

    parts = salary.split('-')

    return int(parts[0]), int(parts[1])

df[['salary_min', 'salary_max']]= df['Salary Estimate'].apply(lambda x: pd.Series(clean_salary(x)))
```

```python
df["Company Name"]=df["Company Name"].str.split('\n').str[0].str.strip()
df["State"]=df["Location"].str.split(',').str[-1].str.strip() 
df["City"]=df["Location"].str.split(',').str[0].str.strip() 
df.drop_duplicates(inplace=True)
df.dropna(subset=["Job Title","Company Name"],inplace=True)
```

```python
skills_list = ['python', 'sql', 'excel', 'tableau', 'power bi',
               'r', 'machine learning', 'statistics', 'pandas', 'spark']

for skill in skills_list:
    df[skill] = df['Job Description'].str.lower().str.contains(skill).astype(int)


df.to_csv("cleaned_data.csv", index=False)
print("✅ Cleaned data saved!")
    
    
```
