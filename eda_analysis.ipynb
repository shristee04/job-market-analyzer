```python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("cleaned_data.csv")
```

```python
skills_cols=['python', 'sql', 'excel', 'tableau', 'power_bi',
               'r', 'machine_learning', 'statistics', 'pandas', 'spark']

skill_demand=df[skills_cols].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=skill_demand.values, y=skill_demand.index, 
             hue=skill_demand.index, palette='viridis', legend=False)
plt.title("Top 10 Most In-Demand Skills for Data Analysts", fontsize=16)
plt.xlabel("Number of job postings")
plt.ylabel("Skills")
plt.tight_layout()
plt.show()
plt.savefig('Top_skills.png')

```

```python
top_titles=df['job_title'].value_counts().head(8).index
salary_df=df[df['job_title'].isin(top_titles)]
plt.figure(figsize=(14,8))
sns.boxplot(data=salary_df, x='job_title', y='salary_avg', hue='job_title', palette='coolwarm', legend=False)
plt.xticks(rotation=45)
plt.title('Salary Distribution by Job Title', fontsize=16)
plt.tight_layout()
plt.show()
```
